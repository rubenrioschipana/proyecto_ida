# 2. Antecedentes del proyecto

## 2.1. Descripción y origen del proyecto

El presente proyecto, denominado **"Ampliación y Mejoramiento de los Sistemas de Agua Potable y Alcantarillado de la Localidad de Cascas"**, surge como respuesta a la necesidad de mejorar las condiciones de abastecimiento de agua potable y saneamiento básico de la población del distrito de Cascas, provincia de Gran Chimú, departamento de La Libertad.

La formulación del proyecto se desarrolla en el marco de los lineamientos establecidos por el Ministerio de Vivienda, Construcción y Saneamiento, así como de las políticas nacionales orientadas a garantizar el acceso adecuado a los servicios básicos de saneamiento y salud pública. Asimismo, el proyecto toma como base los estudios previos de preinversión realizados por la Municipalidad Provincial de Gran Chimú y los documentos técnicos elaborados para el expediente técnico correspondiente.

## 2.2. Problemática existente

Actualmente, el sistema de abastecimiento de agua potable de la localidad de Cascas presenta diversas deficiencias que afectan directamente la continuidad, calidad y cobertura del servicio. Parte importante de la infraestructura hidráulica tiene una antigüedad superior a 40 años, encontrándose en condiciones obsoletas y con frecuentes pérdidas de agua debido al deterioro de las tuberías.

El sistema existente está conformado por dos captaciones principales, líneas de conducción, reservorios y redes de distribución; no obstante, gran parte de las tuberías son de asbesto cemento y presentan deterioro estructural considerable. Asimismo, el sistema no cuenta con un adecuado proceso de desinfección ni potabilización del agua, lo cual representa un riesgo para la salud pública.

De acuerdo con los estudios previos realizados, el servicio de agua potable presenta discontinuidad en el suministro, cobertura insuficiente y elevados porcentajes de pérdidas en la red de distribución. Además, parte de la población no cuenta con conexiones domiciliarias adecuadas.

En relación con el sistema de alcantarillado sanitario, se identificó que la infraestructura existente también presenta problemas asociados a su antigüedad y deficiente operación. El sistema carece de una planta de tratamiento de aguas residuales, ocasionando que las descargas sean vertidas directamente hacia quebradas y terrenos agrícolas, generando contaminación ambiental y riesgos sanitarios para la población.

## 2.3. Situación socioeconómica y necesidad del proyecto

La actividad económica predominante en la localidad de Cascas es la agricultura y ganadería, destacando especialmente la producción vitivinícola artesanal. Debido a ello, la disponibilidad de servicios básicos adecuados constituye un elemento fundamental para mejorar las condiciones de vida y promover el desarrollo económico de la zona.

La deficiente infraestructura sanitaria existente ha generado problemas de salubridad y enfermedades de origen hídrico, principalmente enfermedades gastrointestinales y dermatológicas. Asimismo, la carencia de un sistema adecuado de tratamiento de aguas residuales ha contribuido a la contaminación de terrenos agrícolas y recursos hídricos cercanos.

Frente a esta problemática, las autoridades locales y la población beneficiaria priorizaron la ejecución del presente proyecto con el objetivo de mejorar la cobertura y calidad de los servicios de agua potable y alcantarillado, reducir los riesgos sanitarios y contribuir al bienestar de la población.

## 2.4. Estudios técnicos desarrollados

Para la elaboración del presente proyecto se realizaron diversos estudios básicos que permitieron definir las características técnicas y condiciones del área de intervención. Entre los principales se tienen: el estudio topográfico del área, el estudio de mecánica de suelos, el estudio de impacto ambiental, el catastro técnico de las redes existentes y la evaluación de la infraestructura sanitaria actual.

Estos estudios permitieron obtener información relacionada con la topografía del terreno, las características geotécnicas del suelo, las condiciones ambientales y el estado actual de las redes e infraestructura existente, sirviendo como base para el planteamiento de las soluciones técnicas propuestas.

## 2.5. Alcance general del proyecto

El proyecto contempla la ampliación y mejoramiento integral de los sistemas de agua potable y alcantarillado sanitario de la localidad de Cascas, incluyendo componentes de captación, conducción, almacenamiento, distribución y tratamiento. Entre las principales obras consideradas se encuentran la construcción y mejoramiento de captaciones, la implementación de líneas de conducción, la construcción de reservorios, la renovación y ampliación de redes de distribución, la construcción de redes de alcantarillado sanitario, la implementación de conexiones domiciliarias y la construcción de infraestructura para tratamiento de aguas residuales.

El desarrollo de estas obras permitirá optimizar las condiciones del servicio, incrementar la continuidad del abastecimiento de agua potable y mejorar las condiciones sanitarias y ambientales de la población beneficiaria.

## 2.6. Justificación del proyecto

La ejecución del presente proyecto se justifica debido a la necesidad de reducir los niveles de insalubridad y mejorar las condiciones de vida de la población de Cascas mediante la implementación de infraestructura sanitaria adecuada y sostenible.

El proyecto permitirá disminuir la incidencia de enfermedades relacionadas con el consumo de agua no potabilizada y la inadecuada disposición de aguas residuales. Asimismo, contribuirá a mejorar la calidad ambiental, optimizar el uso de los recursos hídricos y fortalecer el desarrollo urbano y económico de la localidad. Desde el punto de vista técnico y social, representa una solución integral frente a las deficiencias actuales del sistema de saneamiento.
