# 14. Recomendaciones

**1. Ajuste de diámetros mínimos en la red de distribución.** Los tramos identificados en la modelación EPANET con diámetro comercial de 50 mm deben ajustarse a un diámetro mínimo de 75 mm, en cumplimiento estricto de la norma OS.050, aun cuando ello implique sobredimensionar dichos tramos respecto a su requerimiento hidráulico estrictamente calculado. Este ajuste es indispensable para garantizar la seguridad operativa y contra incendio de la red, así como para reducir el riesgo de obstrucciones por sedimentos e incrustaciones a lo largo de su vida útil.

**2. Control de la presión estática en el nodo más bajo de la red.** Dado que la presión estática registrada en el punto más bajo de la red (139.66 m.c.a.) excede ampliamente el límite normativo de 50 m.c.a., se recomienda evaluar en la siguiente etapa de diseño una o varias de las siguientes soluciones técnicas: (i) sectorización de la red en zonas de presión independientes según su cota topográfica; (ii) instalación de Válvulas Reductoras de Presión (VRP) en los tramos o nodos críticos; y (iii) implementación de cámaras o tanques rompecarga en puntos estratégicos del trazado que permitan reiniciar la carga hidráulica disponible en los tramos subsiguientes.

**3. Revisión de tramos con velocidades por debajo del mínimo normativo.** Para los ramales terminales y sectores de baja demanda donde la velocidad no alcanza el mínimo de 0.6 m/s, se recomienda evaluar la interconexión de dichos ramales con tramos aguas abajo (cierre de circuitos) allí donde sea factible, así como incorporar en el plan de operación y mantenimiento purgas periódicas programadas en los puntos de menor velocidad, para prevenir la sedimentación y el estancamiento del agua.

**4. Protección de las líneas de conducción frente a la erosión y los movimientos de tierra.** Considerando la experiencia de deterioro de la infraestructura existente por exposición de tuberías, se recomienda verificar en campo, previamente a la construcción, los sectores de mayor pendiente del trazado de las líneas de conducción LC1 y LC2, asegurando un enterrado con recubrimiento mínimo adecuado y, de ser necesario, protecciones adicionales (dados de anclaje, revestimiento) en los puntos más críticos identificados mediante el estudio de suelos.

**5. Mantenimiento preventivo de las obras de captación.** Se recomienda implementar de manera estricta el protocolo de inspección mensual del cerco perimetral y las cunetas de coronación, así como la purga semestral (o inmediatamente después de episodios de lluvia intensa) de las cámaras húmedas de las captaciones, conforme a lo establecido en el diseño de la captación de fondo de Chimbiles, extendiendo este criterio a la captación de La Peña.

**6. Programa de monitoreo de la calidad del agua.** Se recomienda implementar el Programa de Monitoreo establecido en el Plan de Manejo Ambiental, con una frecuencia mensual inicial (pudiendo pasar a trimestral una vez verificada la buena calidad del agua), midiendo como mínimo turbidez, color, cloro residual, pH, conductividad, dureza total, cloruros, sulfatos y coliformes totales y fecales, con puntos de muestreo en la salida del reservorio y en un caño representativo de la red.

**7. Ajuste operativo de la dosificación de coagulante y cloro.** Se recomienda que el operador de la PTAP realice pruebas de jarras (Jar Test) de forma diaria, especialmente durante la temporada de lluvias, para ajustar la dosis de sulfato de aluminio en función de las variaciones de turbidez de la fuente, y que se mantenga un residual de cloro libre entre 0.30 y 0.60 mg/L a la salida del sistema, conforme a lo establecido en el EIA.

**8. Gestión de lodos y materiales de purga.** Se recomienda implementar los lechos de secado anexos a la PTAP para la deshidratación de los lodos fisicoquímicos generados en el sedimentador, así como disponer el material excedente y los residuos de obra (restos de concreto, tubería, envases) en las trincheras y áreas de depósito de material excedente previstas, evitando su vertimiento en cauces o terrenos agrícolas.

**9. Verificación de los caudales de estiaje de las fuentes.** La Memoria Descriptiva reporta rendimientos históricos de estiaje de 6 L/s para Los Chimbiles y 10 L/s para La Peña, cifras que deben verificarse y actualizarse mediante aforos recientes antes de la etapa de expediente técnico definitivo, a fin de confirmar que los caudales de diseño adoptados (5.36 L/s y 10.00 L/s, respectivamente) cuentan con el respaldo hidrogeológico correspondiente y la licencia de uso de agua vigente ante la ANA.

**10. Cumplimiento del diseño sismorresistente.** Se recomienda que el expediente técnico definitivo incluya la verificación estructural explícita de las cámaras de captación, el reservorio y demás estructuras de concreto armado bajo los criterios de la norma E.030, dado que el proyecto se ubica en una región de actividad sísmica moderada a alta.

**11. Señalización y control de tiempos en la etapa de construcción.** Conforme a lo señalado en el EIA, se recomienda que las zanjas para la instalación de tuberías no permanezcan abiertas por más de 24 horas y se mantengan debidamente señalizadas, minimizando así el riesgo de accidentes para la población y el personal de obra, así como la afectación temporal del tránsito peatonal y vehicular.

**12. Fortalecimiento de capacidades locales.** Se recomienda complementar la ejecución de la obra con un programa de capacitación dirigido a la Junta Administradora de Servicios de Saneamiento (JASS) u operador responsable, que incluya el manejo del hipoclorador, la interpretación de los registros de la Bitácora de Operación, el mantenimiento de válvulas y accesorios, y los protocolos de purga y limpieza, a fin de garantizar la sostenibilidad operativa del sistema en el largo plazo.

**13. Actualización del catastro y seguimiento post-construcción.** Finalmente, se recomienda mantener actualizado el catastro técnico de las redes una vez ejecutada la obra, y realizar un seguimiento del comportamiento hidráulico real del sistema durante los primeros años de operación, contrastándolo con los resultados de la modelación en EPANET, de modo que cualquier desviación respecto a lo proyectado pueda corregirse oportunamente.
