# 12. Descripción de riesgos naturales asociados al sistema

## 12.1. Marco general

El sistema de abastecimiento de agua potable proyectado para la localidad de Cascas se desarrolla íntegramente en un entorno de sierra andina, con una topografía accidentada que va desde los 1 505.739 m s. n. m. en la captación La Peña hasta los 1 373.165 m s. n. m. en el punto de entrega a la red, y con un clima cálido seco de precipitación pluvial media-alta. Estas condiciones físicas y climáticas determinan los principales riesgos naturales que deben ser considerados en el diseño, construcción y operación de la infraestructura, los cuales se identifican a partir de la información recogida en el Estudio de Impacto Ambiental (EIA) y en la Memoria Descriptiva del proyecto.

## 12.2. Riesgo hidrometeorológico: precipitaciones y erosión

La zona presenta una **precipitación pluvial media-alta** dentro de un clima cálido seco, con temperaturas que oscilan entre 12 °C y 19 °C. Este régimen de lluvias estacionales representa el principal factor de riesgo natural para el sistema, por dos vías:

- **Erosión y arrastre de sedimentos hacia las captaciones.** Las precipitaciones intensas generan escorrentía superficial que puede arrastrar sedimentos, materia orgánica y contaminantes hacia el punto de afloramiento de las fuentes (La Peña y Los Chimbiles), afectando la calidad del agua captada. Por ello, el diseño de la captación de fondo de Chimbiles incorpora cunetas de coronación perimetrales cuya inspección y limpieza mensual es obligatoria, evitando que las aguas de escorrentía pluvial ingresen al afloramiento y contaminen la fuente.
- **Exposición y deterioro de tuberías.** La Memoria Descriptiva señala que buena parte de la infraestructura existente en Cascas (línea de conducción y red de distribución) se encuentra con tramos de tubería expuestos debido a la erosión del terreno y a la antigüedad de las instalaciones (más de 40 años en varios sectores), lo cual constituye un riesgo de rotura por maltrato o socavación. El proyecto de ampliación y mejoramiento contempla el reemplazo íntegro de estos tramos, debiendo mantenerse un enterrado adecuado de las nuevas líneas de conducción y aducción para mitigar este riesgo.

## 12.3. Riesgo de movimientos de tierra en laderas

Dado que las líneas de conducción (LC1 y LC2) discurren por laderas de fuerte pendiente propias del relieve andino —con desniveles de más de 100 m entre las captaciones y el reservorio—, existe un riesgo latente de deslizamientos o derrumbes que podrían afectar la zanja de instalación de las tuberías, especialmente durante o después de eventos de lluvia intensa. Este riesgo es reconocido de forma indirecta en el EIA, que identifica los movimientos de tierra (apertura y tapado de zanjas) como una de las actividades de mayor potencial de impacto —y de exposición al riesgo— tanto para el componente suelo como para la seguridad del personal de obra, recomendando que el tiempo de zanjas abiertas no exceda las 24 horas y que estén debidamente señalizadas.

## 12.4. Riesgo sísmico

Si bien ni el EIA ni la Memoria Descriptiva desarrollan un estudio sismológico específico, el diseño estructural de los componentes principales del sistema (reservorio de 300 m³, cámaras de captación, muros de contención) se rige por la normativa nacional de concreto armado (RNE E.060) y debe además satisfacer los criterios sismorresistentes de la norma E.030, en tanto el Perú se ubica en una región de alta actividad sísmica. Este criterio ha sido incorporado en los diseños de las estructuras de captación y almacenamiento del sistema.

## 12.5. Riesgo de contaminación de la fuente por actividad antrópica

Más que un riesgo estrictamente natural, el EIA identifica un riesgo mixto (natural-antrópico) relevante para la seguridad hídrica del sistema: la posible alteración de la calidad del agua de la fuente por vertimientos accidentales de residuos de construcción (concreto, cemento) durante la etapa de obra, cuyos efectos se manifestarían aguas abajo de la captación. Aunque calificado como de magnitud pequeña y temporal, este riesgo refuerza la importancia de las medidas de protección física de las captaciones (cerco perimetral de malla olímpica, tapa sanitaria hermética) contempladas en el diseño de la captación de fondo.

## 12.6. Medidas de mitigación incorporadas al diseño

En conjunto, el sistema incorpora las siguientes medidas de mitigación frente a los riesgos naturales identificados:

- Cunetas de coronación y cerco perimetral en las captaciones, con inspección y limpieza periódica.
- Tapa sanitaria hermética que evita el ingreso de agentes externos y la incidencia de radiación solar (crecimiento de algas).
- Diseño estructural de reservorio, captaciones y cámaras conforme a las normas E.060 (concreto armado) y criterios sismorresistentes.
- Enterrado adecuado de las líneas de conducción y aducción para reducir la exposición a la erosión superficial y a eventos de deslizamiento.
- Protocolos de purga semestral de las cámaras de captación, reforzados tras episodios de lluvias intensas.
- Plan de Manejo Ambiental y Programa de Monitoreo de la calidad del agua (turbidez, coliformes, cloro residual, entre otros parámetros) que permite detectar oportunamente cualquier afectación de la fuente asociada a eventos hidrometeorológicos o antrópicos.

Estas medidas, aplicadas de manera conjunta, reducen a niveles aceptables la exposición del sistema a los riesgos naturales propios de su emplazamiento en la sierra andina, sin que se identifiquen condiciones que comprometan la viabilidad técnica del proyecto.
