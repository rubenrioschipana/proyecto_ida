
# 1. Generalidades

## 1.1. Objetivo del proyecto

El presente informe tiene por objetivo desarrollar, a nivel de perfil técnico, el diseño integral del sistema de abastecimiento de agua potable de la localidad de Cascas, abarcando todos sus componentes: captación, conducción, tratamiento, almacenamiento, aducción y distribución. La finalidad es plantear una solución que garantice el suministro adecuado, continuo y de calidad para la población beneficiaria durante el horizonte de diseño previsto.

## 1.2. Ubicación

El proyecto se localiza en la **localidad de Cascas, distrito de Cascas, provincia de Gran Chimú, departamento de La Libertad**, en la vertiente occidental de los Andes, a una altitud aproximada de 1 279 m s. n. m. La zona presenta un clima cálido seco durante todo el año, con temperaturas que oscilan entre 12 °C y 19 °C, y una economía predominantemente agrícola y de transformación en la que destaca la actividad pecuaria asociada a la producción de leche.

## 1.3. Marco normativo

El diseño se desarrolla conforme al **Reglamento Nacional de Edificaciones (RNE)** y la normativa sectorial vigente, en particular:

- **OS.010** — Captación y conducción de agua para consumo humano.
- **OS.020** — Plantas de tratamiento de agua para consumo humano.
- **OS.030** — Almacenamiento de agua para consumo humano.
- **OS.050** — Redes de distribución de agua para consumo humano.
- **OS.100** — Consideraciones básicas de diseño de infraestructura sanitaria.
- **D.S. N.° 031-2010-SA** — Reglamento de la Calidad del Agua para Consumo Humano.
- **Ley N.° 29338** — Ley de Recursos Hídricos (licencia de uso de agua otorgada por la ANA).

## 1.4. Período de diseño

Se adopta un **período de diseño de 20 años**, justificado por tratarse de una población joven y en proceso de desarrollo, con una tendencia de crecimiento sostenida. Un horizonte de 20 años permite amortizar adecuadamente la inversión sin incurrir en un sobredimensionamiento excesivo en las etapas iniciales de operación, equilibrando el costo de la obra con la vida útil de los materiales (tuberías de PVC) y la incertidumbre asociada a la proyección demográfica.

## 1.5. Filosofía de diseño

Dada la configuración topográfica de la zona y la ubicación de las fuentes de captación en la parte alta de la cuenca, el sistema ha sido concebido para operar íntegramente **por gravedad**. Este enfoque elimina la necesidad de sistemas de bombeo, reduciendo significativamente los costos operativos y de mantenimiento electromecánico para la junta administradora local, y aprovecha la energía potencial del relieve andino tanto para la conducción como para los procesos de tratamiento.
