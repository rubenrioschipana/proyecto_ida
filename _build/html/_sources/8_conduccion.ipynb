
# 8. Diseño de la línea de conducción 1 y 2 (La Peña y Los Chimbiles) hacia el reservorio

La línea de conducción transporta el agua desde la obra de captación hasta el reservorio de almacenamiento. Su diseño busca garantizar que el caudal de diseño ($Q_{md}$) llegue al reservorio aprovechando la energía disponible por gravedad, sin que las presiones superen la capacidad resistente de la tubería ni las velocidades salgan del rango admisible. La conducción se resuelve mediante **dos líneas independientes**: **LC1**, proveniente de la captación **La Peña / Platanar** (10.00 L/s), y **LC2**, proveniente de la captación **Los Chimbiles** (5.36 L/s). En conjunto transportan el caudal máximo diario hacia el reservorio: $10.00 + 5.36 = 15.36 \ \text{L/s} \approx Q_{md}$.

## 8.1. Criterios de diseño

Ambas líneas se proyectan en **tubería de PVC**, material de baja rugosidad, buena resistencia a la presión y costo competitivo. Los criterios aplicados son:

**1. Carga estática disponible ($H_t$).** Diferencia de cotas entre el inicio (captación) y la cota más baja del trazo; representa la energía potencial máxima disponible para vencer las pérdidas y entregar el agua al reservorio:

$$ H_{t} = Z_{inicio} - Z_{m\acute{i}n} $$

**2. Presión de diseño ($P_d$).** La condición más exigente ocurre cuando el flujo se detiene (condición estática). Para seleccionar la clase de tubería se afecta la carga estática máxima por un factor de seguridad de 1.2:

$$ P_{d} = 1.2 \cdot H_{t} $$

**3. Velocidades admisibles (RNE).** $V_{m\acute{i}n} = 0.6$ m/s y $V_{m\acute{a}x} = 3.0$ m/s para PVC. El límite inferior evita la **sedimentación**; el superior previene la **erosión** y limita el golpe de ariete. El diámetro teórico se obtiene de la ecuación de continuidad:

$$ D = \sqrt{\dfrac{4\,Q}{\pi\,V}} $$

**4. Pérdida de carga (Hazen-Williams).** Con coeficiente $C = 150$ para PVC:

$$ h_{f} = \dfrac{10.674 \cdot L \cdot Q^{1.852}}{C^{1.852} \cdot D^{4.871}} $$

El diseño es satisfactorio cuando la pérdida de carga total es menor que la carga disponible ($h_{f} < H_{t}$) y la velocidad se mantiene dentro del rango admisible.

## 8.3. Línea de conducción LC1 (La Peña / Platanar)

La línea **LC1** es la de mayor longitud y caudal del sistema. Transporta **10.00 L/s** desde la cota 1 505.739 m s. n. m. hasta el reservorio (1 390.747 m s. n. m.), a lo largo de **2 937.96 m**. La cota más baja del recorrido (1 362.958 m s. n. m.) define una carga estática disponible de $H_{t} = 114.99$ m y una presión de diseño de $P_{d} = 171.34$ mca (17.48 bar), por lo que se adopta **tubería PVC clase 20 (PN20)**, cuya presión nominal de 20 bar resiste con holgura la presión de diseño.

El rango de diámetros teóricos para las velocidades admisibles es de 2.56″ a 5.74″; se adopta el diámetro comercial **3½″ (88.9 mm)**. Con $C = 150$ y la longitud del trazo, la pérdida de carga resulta $h_{f} = 76.21$ m, inferior a la carga disponible, por lo que existe energía suficiente para conducir el agua al reservorio. La velocidad es $V = 1.61$ m/s, dentro del rango admisible. La línea, por tanto, **cumple**.

La presión dinámica máxima a lo largo de LC1 es del orden de 70.6 mca, muy inferior a la presión nominal de la tubería, por lo que la conducción opera con un amplio margen de seguridad estructural.

## 8.4. Línea de conducción LC2 (Los Chimbiles)

La línea **LC2** transporta **5.36 L/s** desde la cota 1 481.059 m s. n. m. hasta la cota 1 390.983 m s. n. m., a lo largo de **1 206.11 m**. La carga estática disponible es $H_{t} = 90.08$ m, de la que resulta una presión de diseño de $P_{d} = 108.09$ mca (11.03 bar); se adopta **tubería PVC clase 12.5 (PN12.5)**, solución más económica adecuada a la menor presión.

El rango de diámetros teóricos es de 1.88″ a 4.20″; se adopta el diámetro comercial **3″ (76.2 mm)**. Con $C = 150$ y la longitud del trazo, la pérdida de carga es $h_{f} = 20.89$ m, muy por debajo de la carga disponible, y la velocidad es $V = 1.18$ m/s, admisible. La línea **cumple**.

La L.G.H. se mantiene por encima del terreno en todo el recorrido, lo que confirma que no existen presiones negativas que pudieran provocar el ingreso de aire o el colapso de la tubería.

El cálculo de la línea de gradiente hidráulica, desarrollado tramo a tramo para las 61 progresivas de LC1 y las 16 de LC2, confirma que las presiones dinámicas máximas quedan por debajo de la presión nominal de la clase de tubería seleccionada en cada caso. Ambas líneas satisfacen simultáneamente las exigencias hidráulicas (energía suficiente y velocidades admisibles) y estructurales (resistencia a la presión).
