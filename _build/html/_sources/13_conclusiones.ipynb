# 13. Conclusiones

**1. Sobre la población y los caudales de diseño.** Se adoptó una población de diseño de **6 807 habitantes** al horizonte de 20 años (año 2046), determinada mediante el modelo de proyección polinómico de segundo grado, por presentar el mejor coeficiente de determinación (R² = 0.886) y ser el más coherente con la tendencia de densificación urbana asociada a la producción industrial de leche en la localidad. Con una dotación de diseño de 150 L/hab/d (criterio SEDAPAL, nivel socioeconómico D), se obtuvieron los caudales característicos del sistema: Qm = 11.82 L/s, Qmd = 15.36 L/s y Qmh = 23.64 L/s, no siendo necesario considerar caudal contra incendio por tratarse de una población menor a 10 000 habitantes.

**2. Sobre la captación.** El sistema se abastece de dos fuentes independientes: la captación **La Peña / Platanar** (10.00 L/s) y la captación de manantial de fondo **Los Chimbiles** (5.36 L/s), cuya suma (15.36 L/s) coincide con el caudal máximo diario de diseño. La captación de Chimbiles se proyectó como una cámara de concreto armado sin losa de fondo, con cámara húmeda y seca diferenciadas, filtro de grava, tapa sanitaria hermética y cerco perimetral, en estricto cumplimiento de la norma RNE OS.010.

**3. Sobre las líneas de conducción.** Las líneas LC1 (La Peña, 2 937.96 m) y LC2 (Los Chimbiles, 1 206.11 m) transportan el agua hasta el reservorio íntegramente por gravedad. Ambas líneas cumplen simultáneamente los criterios hidráulicos (Hf < Ht, velocidades entre 0.6 y 3.0 m/s) y estructurales (presión dinámica máxima muy por debajo de la presión nominal de la clase de tubería adoptada: clase 20 para LC1 y clase 12.5 para LC2).

**4. Sobre la planta de tratamiento.** Dada la vulnerabilidad hidroclimática de las fuentes superficiales de régimen lótico ante eventos de lluvia intensa, se proyectó una **Planta de Tratamiento Convencional de Filtración Rápida** (desarenador, mezcla rápida en canal Parshall, floculador hidráulico de pantallas, sedimentador de alta tasa con módulos de placas, filtración rápida de lecho dual y desinfección con hipoclorito de calcio), conforme a la Norma OS.020 y a los manuales del CEPIS/OPS, garantizando una turbidez de salida menor a 1.0 UNT.

**5. Sobre el reservorio.** Se adoptó un reservorio apoyado de sección circular en concreto armado, con una capacidad de **300 m³**, superior al volumen mínimo de regulación normativo (255.31 m³, equivalente al 25 % del volumen promedio diario), lo que otorga un margen adicional de resiliencia operativa. Las tuberías de entrada, salida, limpieza y rebose se estandarizaron en su mayoría en PVC de 6″, y se incorporó un sistema de dosificación continua de hipoclorito de calcio (3.06 kg/día) para la desinfección final.

**6. Sobre la línea de aducción.** El tramo de 115.00 m que conecta el reservorio con la red de distribución, dimensionado con el caudal máximo horario (23.64 L/s), cumple ampliamente los criterios hidráulicos y estructurales, con una pérdida de carga (1.06 m) muy inferior a la carga estática disponible (25.48 m) y una tubería de clase 5, la de menor presión nominal del sistema.

**7. Sobre la red de distribución.** La modelación hidráulica en EPANET V2.2, con demandas asignadas mediante el método de los Polígonos de Thiessen, evidencia que la red cumple de manera general con los parámetros normativos de la OS.050. No obstante, se identifican tres condiciones puntuales que requieren atención antes de la etapa de ejecución: (i) tramos con velocidades por debajo del mínimo normativo, asociados a sectores de baja demanda o ramales terminales; (ii) tramos cuyo diámetro hidráulicamente requerido (50 mm) resulta inferior al mínimo normativo de 75 mm exigido por razones de seguridad operativa y contra incendio; y (iii) una presión estática de 139.66 m en el nodo más bajo de la red, muy superior al límite normativo de 50 m.c.a., consecuencia directa del marcado desnivel topográfico entre el reservorio y la zona baja de la localidad.

**8. Sobre los riesgos naturales.** El emplazamiento del sistema en un entorno de sierra andina, con pendientes pronunciadas y precipitación pluvial media-alta, expone a la infraestructura a riesgos de erosión, arrastre de sedimentos hacia las captaciones y eventuales movimientos de tierra en las laderas de trazo de las líneas de conducción. Estos riesgos han sido considerados en el diseño mediante medidas de protección física de las captaciones, criterios sismorresistentes en las estructuras de concreto armado y un programa de monitoreo de la calidad del agua, por lo que no se identifican condiciones que comprometan la viabilidad técnica del proyecto.

**9. Sobre el impacto ambiental.** De acuerdo con el Estudio de Impacto Ambiental, los impactos ambientales potenciales de mayor relevancia son los **positivos**, concentrados en la etapa de funcionamiento del sistema (mejora de la salud pública, reducción de enfermedades gastrointestinales, generación de empleo). Los impactos negativos se concentran en la etapa de construcción (movimientos de tierra, alteración temporal de la calidad del aire, suelo y paisaje), son en su mayoría de baja a moderada magnitud y cuentan con alta posibilidad de aplicación de medidas de prevención y mitigación, lo que hace viable la ejecución de la obra.

**10. Conclusión general.** El sistema de abastecimiento de agua potable proyectado para Cascas —captación, conducción, tratamiento, almacenamiento, aducción y distribución— constituye una solución técnicamente viable e integralmente concebida para operar por gravedad, sin necesidad de sistemas de bombeo, aprovechando la energía potencial del relieve andino en todas sus etapas. El diseño satisface los criterios hidráulicos y normativos exigidos por el Reglamento Nacional de Edificaciones (OS.010, OS.020, OS.030 y OS.050), quedando pendientes únicamente los ajustes puntuales identificados en la red de distribución, que se abordan en el capítulo de recomendaciones.
