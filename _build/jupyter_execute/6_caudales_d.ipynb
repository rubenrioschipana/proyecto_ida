
# 6. Caudales de diseño

## 6.1. Dotación

La **dotación** es el volumen de agua asignado por habitante y por día y, junto con la población, es el segundo parámetro determinante de la demanda. Su valor depende del clima, los hábitos de consumo, el nivel socioeconómico y los usos previstos. La localidad presenta un **clima cálido seco**, con temperaturas entre 12 °C y 19 °C; económicamente predomina una economía de subsistencia, con un ingreso promedio mensual del orden de S/ 550 en el área urbana (nivel socioeconómico D).

Se evaluaron dos referencias normativas:

| Referencia                                        | Dotación     |
|:--------------------------------------------------|:------------:|
| RNE OS.100 (clima cálido seco, lotes > 90 m²)     | 180 L/hab/d  |
| SEDAPAL (nivel socioeconómico D)                  | 150 L/hab/d  |

*Tabla 6.1. Dotación según referencia normativa.*

El RNE, atendiendo únicamente al clima y al tamaño de lote, asignaría 180 L/hab/d; sin embargo, este valor no recoge el nivel de consumo efectivo de una población de economía de subsistencia. El criterio de SEDAPAL, que sí incorpora el nivel socioeconómico, propone 150 L/hab/d, valor más representativo del consumo real. Por ello se adopta una **dotación de diseño de 150 L/hab/d**.

## 6.2. Caudales de diseño

El consumo no es uniforme en el tiempo: varía a lo largo del año (días de mayor demanda) y del día (horas pico). Para que cada estructura funcione adecuadamente en su condición más desfavorable se definen tres caudales característicos.

**Caudal promedio diario ($Q_m$).** Consumo medio de la población a lo largo de un año:

$$ Q_{m} = \dfrac{P \cdot Dot}{86\,400} = \dfrac{6\,807 \times 150}{86\,400} = 11.82 \ \text{L/s} $$

**Caudal máximo diario ($Q_{md}$).** Día de mayor consumo del año ($K_{1} = 1.3$):

$$ Q_{md} = K_{1} \cdot Q_{m} = 1.3 \times 11.82 = 15.36 \ \text{L/s} $$

**Caudal máximo horario ($Q_{mh}$).** Hora de mayor consumo del día más desfavorable ($K_{2} = 2.0$):

$$ Q_{mh} = K_{2} \cdot Q_{m} = 2.0 \times 11.82 = 23.64 \ \text{L/s} $$

| Caudal              | Coeficiente | Valor      |
|:--------------------|:-----------:|:----------:|
| Promedio diario, Qm | —           | 11.82 L/s  |
| Máximo diario, Qmd  | K₁ = 1.3    | 15.36 L/s  |
| Máximo horario, Qmh | K₂ = 2.0    | 23.64 L/s  |

## 6.3. Asignación del caudal por tipo de estructura

El caudal con que se dimensiona cada estructura depende de su función. Las estructuras **aguas arriba del tanque** (captación, conducción, planta de tratamiento, aducción y el propio tanque) operan de forma continua y se diseñan con el **caudal máximo diario ($Q_{md}$)**, ya que el tanque amortigua las variaciones horarias. Las estructuras **aguas abajo del tanque** (línea de alimentación y red de distribución) atienden directamente la demanda instantánea y se diseñan con el **caudal máximo horario ($Q_{mh}$)**.

| Estructura                       | Caudal de diseño |
|:---------------------------------|:----------------:|
| Obra de captación                | Qmd              |
| Línea de conducción              | Qmd              |
| Planta de tratamiento            | Qmd              |
| Línea de aducción                | Qmd              |
| Tanque de almacenamiento         | Qmd              |
| Línea de alimentación de la red  | Qmh              |
| Red de distribución              | Qmh              |
