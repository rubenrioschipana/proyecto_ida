# 5. Población de diseño

## 5.1. Información censal

La determinación de la población futura constituye el punto de partida del dimensionamiento, pues define la magnitud de la demanda que se propaga a todas las estructuras del sistema. La estimación parte de la información oficial del INEI, correspondiente a los censos nacionales de 1993, 2005, 2007 y 2017. Dado que el sistema atiende exclusivamente al área urbana, el análisis se desarrolla sobre la **población urbana** de cada censo, descartando la componente rural, abastecida por otros sistemas. La serie histórica evidencia el carácter mayoritariamente rural del distrito —la población urbana representa entre el 29 % y el 32 % del total— pero también una tendencia de urbanización progresiva.

## 5.2. Factor de mayoración

Los censos suelen presentar un cierto grado de **subregistro**, particularmente en zonas dispersas. Para corregir este efecto y trabajar del lado de la seguridad, por indicación del docente (grupo 1) se aplica a la población urbana censal un **factor de mayoración** $K = 1.15$:

$$ P_{corr} = K \cdot P_{censo}, \qquad K = 1.15 $$

## 5.3. Métodos de proyección de la población

A partir de la serie corregida se proyecta la población al horizonte de diseño. Dado que ningún método es universalmente válido, se ajustaron **cinco modelos de crecimiento** y se compararon entre sí. Sea $t$ el año y $P(t)$ la población proyectada:

- **Lineal —** $P(t) = a + b\,t$. Crecimiento por incrementos absolutos constantes; representa poblaciones estabilizadas o de crecimiento moderado.
- **Potencial —** $P(t) = a\,t^{\,b}$. Crecimiento a tasa decreciente, útil cuando el ritmo de incremento se atenúa con el tiempo.
- **Exponencial —** $P(t) = a\,e^{\,b t}$. Tasa de crecimiento porcentual constante; tiende a sobreestimar en horizontes largos.
- **Logarítmico —** $P(t) = a + b\,\ln(t)$. Crecimiento que se desacelera marcadamente, propio de poblaciones próximas a la saturación.
- **Polinómico de 2.º grado —** $P(t) = a + b\,t + c\,t^{2}$. Permite capturar una aceleración (o desaceleración) del crecimiento, adaptándose a tendencias no lineales.

Cada modelo se ajusta por mínimos cuadrados a los cuatro datos censales corregidos y se evalúa mediante el **coeficiente de determinación** $R^{2}$ (un valor cercano a 1 indica un mejor ajuste). Los modelos se proyectan al año 0 (2026), a 10 años (2036) y a 20 años (2046).

| Modelo          | 2026  | 2036  | 2046  | $R^{2}$ |
|:----------------|:-----:|:-----:|:-----:|:-------:|
| Lineal          | 5 582 | 5 781 | 5 980 | 0.8478  |
| Potencial       | 5 591 | 5 807 | 6 030 | 0.8484  |
| Exponencial     | 5 593 | 5 812 | 6 040 | 0.8491  |
| Logarítmico     | 5 580 | 5 776 | 5 972 | 0.8470  |
| **Polinómico G2** | **5 773** | **6 239** | **6 807** | **0.8859** |

La Figura 5.1 evidencia un comportamiento característico: los modelos lineal, potencial, exponencial y logarítmico arrojan proyecciones prácticamente coincidentes entre sí —dado que en el rango de datos analizado todos se aproximan a una recta— y predicen una población del orden de 6 000 hab al año 2046. ## 5.4. Selección del modelo y población de diseño

La selección se fundamenta en dos criterios complementarios. Desde el punto de vista estadístico, el **modelo polinómico de segundo grado** es el que mejor representa la serie histórica, con el mayor coeficiente de determinación ($R^{2} = 0.886$) frente al rango de 0.847–0.849 de los demás modelos. Desde el punto de vista del contexto local, es además el más coherente con la tendencia de **densificación** asociada a la consolidación de la producción industrial de leche, actividad que genera empleo, atrae población e incentiva la concentración de habitantes en el área urbana; adoptar un modelo más conservador implicaría el riesgo de que la infraestructura quede rebasada antes de cumplir su período de diseño.

En consecuencia, se adopta la proyección del modelo polinómico al horizonte de 20 años:

$$ \boxed{P_{dise\tilde{n}o} = 6\,807 \ \text{hab} \quad (\text{horizonte } 2046,\ 20 \ \text{años})} $$
