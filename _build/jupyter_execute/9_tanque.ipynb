
# 9. Diseño del tanque de almacenamiento (reservorio)

## 9.1. Generalidades y filosofía de diseño

El presente diseño corresponde a la unidad de almacenamiento del sistema. Conforme a la filosofía de operación por gravedad del proyecto, el reservorio no requiere sistemas de bombeo, lo que reduce significativamente los costos operativos y de mantenimiento electromecánico para la junta administradora local.

## 9.2. Volumen de regulación y marco normativo

El volumen del reservorio tiene la función primordial de compensar las variaciones horarias del consumo a lo largo del día, almacenando agua en los momentos de bajo consumo y liberándola durante las horas punta.

En cumplimiento del **RNE, Norma Técnica OS.030 (Almacenamiento de Agua para Consumo Humano)**, el cálculo del volumen de regulación suele realizarse mediante un diagrama de masa fundamentado en las variaciones horarias. Sin embargo, para localidades donde no se dispone de registros hidrométricos horarios fiables, la normativa exige que el volumen de regulación sea equivalente, como mínimo, al 25 % del volumen promedio diario anual. Con un caudal promedio de diseño ($Q_m$) de 11.82 L/s (equivalente a 1 021.25 m³/día):

$$ V_{reg} = 0.25 \times Q_m \times \left( \frac{86\,400}{1000} \right) = 0.25 \times 1\,021.25 = 255.31 \, \text{m}^3 $$

Para garantizar la resiliencia operativa del sistema, compensar eventuales emergencias y adaptarse a volúmenes comerciales estandarizados, se ha adoptado un reservorio de **300 m³**.

## 9.3. Predimensionamiento geométrico y consideraciones estructurales

Se ha proyectado un reservorio de tipo apoyado y sección transversal circular, construido en concreto armado. La geometría circular distribuye las presiones hidrostáticas de manera uniforme en forma de esfuerzos de tracción anular, optimizando la cuantía de acero requerida.

De acuerdo con las restricciones del área del proyecto y buscando optimizar el área de excavación en plataforma, se ha establecido una altura útil de agua ($h_1$) de 6.50 m. Adicionalmente, la Norma OS.030 exige prever una cámara de aire en la parte superior para permitir la fluctuación libre de la lámina de agua y la ventilación, por lo que se ha asignado un borde libre ($h_2$) de 0.30 m, definiendo una altura total del muro interior de 6.80 m. Con base en el volumen adoptado y la altura útil, las dimensiones internas son:

$$ A = \frac{300}{6.50} = 46.15 \, \text{m}^2 $$

$$ D = \sqrt{\frac{4 \cdot 46.15}{\pi}} = 7.67 \, \text{m} \rightarrow \text{se adopta constructivamente } 7.70 \, \text{m} $$

## 9.4. Dimensionamiento hidráulico de tuberías y accesorios

Las líneas que interactúan con el reservorio se han dimensionado garantizando velocidades de flujo (asumidas en $V = 1.5$ m/s) que eviten la sedimentación o el desgaste por abrasión.

### 9.4.1. Tuberías principales (entrada y salida)

La línea de entrada, encargada de surtir al tanque, se dimensionó con el Caudal Máximo Diario ($Q_{md} = 15.36$ L/s), resultando en un diámetro teórico de 4.50″. La línea de aducción (salida hacia la red) debe suministrar el Caudal Máximo Horario ($Q_{mh} = 23.64$ L/s), requiriendo un diámetro teórico de 5.58″. En busca de una estandarización constructiva que facilite la instalación y la provisión de válvulas de paso, **se ha adoptado tubería de PVC de 6″ para ambas líneas**.

### 9.4.2. Tubería de limpia y drenaje

La Norma OS.030 exige que todo reservorio cuente con una facilidad de drenaje en el fondo, capaz de evacuar el volumen total en un máximo de 2 horas (7 200 s). Aplicando la ecuación de vaciado de Torricelli con un coeficiente de descarga $C_d = 0.60$:

$$ A_t = \frac{2 \cdot 46.15 \cdot \sqrt{6.50}}{0.60 \cdot 7\,200 \cdot \sqrt{19.62}} = 0.0123 \, \text{m}^2 $$

El área calculada arroja un diámetro teórico de 4.93″. Para garantizar el margen de seguridad y estandarizar con las demás líneas, **se adopta una tubería de PVC de 6″**.

### 9.4.3. Tubería de rebose

Como medida de salvaguarda ante fallas en la válvula flotadora de entrada, la tubería de rebose se diseña para evacuar todo el caudal que ingresa ($Q_{md}$). Asumiendo una carga hidráulica sobre la embocadura de 0.05 m, se obtiene un requerimiento teórico de 3.76″. **Se ha adoptado tubería de PVC de 4″**.

### 9.4.4. Canastilla de succión

Con el objeto de retener partículas que pudiesen obstruir la red y reducir la velocidad puntual de succión que podría generar vórtices, se previó una canastilla ranurada en la tubería de salida. Cumpliendo los parámetros normativos, su diámetro se fijó en 12″ ($2 \times D_{salida}$) y su longitud en 1.00 m, ubicándose en el rango exigido de entre 3 a 6 veces su diámetro.

## 9.5. Sistema de desinfección y obras civiles complementarias

Para garantizar la calidad microbiológica del agua distribuida, se ha considerado un dosificador continuo de hipoclorito de calcio (al 65 % de pureza). Con una dosificación preventiva de 1.5 mg/L para el caudal $Q_{md}$, el sistema demandará **3.06 kg/día** de insumo desinfectante.

El diseño integral incorpora además los siguientes componentes de obra civil esenciales para la preservación sanitaria e integridad de la estructura:

- **Pendiente de losa.** Se dispuso una pendiente de fondo de 1.5 % dirigida hacia el sumidero de la tubería de limpia, garantizando la purga absoluta de sedimentos.
- **Ventilación.** Se incluyó una tubería de PVC de 4″ tipo cuello de cisne en la cubierta, con malla metálica milimétrica, para disipar las fluctuaciones de presión sin permitir el ingreso de vectores biológicos.
- **Línea de by-pass.** Se proyectó una conexión externa perimetral en PVC de 6″ que enlaza la línea de conducción con la de aducción, permitiendo mantener abastecida a la red por gravedad durante las horas de mantenimiento del reservorio.
